```python
def signature(word):
    return ''.join(sorted(word))
```

```python
def anagram(myword):
    return [word for word in wordclean if signature(word) == signature(myword)]
```

```python
anagram('dictionary')
```

```python
%timeit anagram('dictionary')
```
